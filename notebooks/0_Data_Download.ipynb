### üìÅ Step 2 ‚Äî Folder Structure and Naming Convention

All downloaded files should be saved in:

`/app/data/raw/sample/`


Each product must be stored in a **separate subfolder** following this naming scheme:

`<Company>_<product_id>/`

Example:

`/app/data/raw/sample/Siemens_6ES71314BF000AA0/`


This structure ensures that each product can be uniquely identified across manufacturers, avoiding potential ID collisions.
